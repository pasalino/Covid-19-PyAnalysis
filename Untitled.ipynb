```python
import os
import subprocess
import pandas as pd 
import datetime
import matplotlib
from matplotlib import pyplot as plt
```

```python
PROTEZIONE_CIVILE_REPO = 'git@github.com:pcm-dpc/COVID-19.git'
DATA_DIR = './data'
REPO_DIR = os.path.join(DATA_DIR,'repo')
```

```python
if not os.path.isdir(REPO_DIR):
    print('### CLONE ###')
    p = subprocess.Popen(['git', 'clone', '--single-branch', '-b', 'master', PROTEZIONE_CIVILE_REPO, REPO_DIR])
    p.wait()
    print('### DONE CLONE ###')
else:
    print('### PULL ###')
    p = subprocess.Popen(['git', 'pull'])
    p.wait()
    print('### DONE PULL ###')
    
```

```python
national_trend = pd.read_csv(os.path.join(REPO_DIR,'dati-andamento-nazionale/dpc-covid19-ita-andamento-nazionale.csv'))
```

```python
national_trend['data'] = pd.to_datetime(national_trend['data']).dt.date
```

```python
incremental_data = pd.DataFrame()
incremental_data['data'] = pd.to_datetime(national_trend['data']).dt.date
incremental_data['nuovi_attualmente_positivi'] = national_trend['nuovi_attualmente_positivi']
```

```python
incremental_columns = ['ricoverati_con_sintomi', 'terapia_intensiva', 'totale_ospedalizzati', 'isolamento_domiciliare', 'totale_attualmente_positivi', 'dimessi_guariti', 'deceduti', 'totale_casi', 'tamponi']

for c in incremental_columns:
    incremental_data[c] = national_trend[c] - national_trend[c].shift(1)
    incremental_data[c] = incremental_data[c].astype(pd.Int64Dtype())

complete_incremental_data = incremental_data
incremental_data = incremental_data[1:]
```

## INCREMENTI GIORNALIERI

```python
# %matplotlib notebook
plt.rcParams['figure.figsize'] = [20, 7]

plot =['nuovi_attualmente_positivi','ricoverati_con_sintomi', 'terapia_intensiva','isolamento_domiciliare','dimessi_guariti','deceduti']
colors=['red','green','#00FF00','#0000FF','#3AFAF0','#000000']
legends = ["Nuovi Positivi","Ricoverati","Terapia Intensiva","Isolamento domiciliare","Guariti","Deceduti"]

ax = incremental_data.plot(kind='line', x='data', y=plot, color=colors, rot=90, label=legends)
ax.legend(loc="upper left")
ax.set_xlabel('Date')
ax.set_ylabel('Casi')
ax.grid(b=True, which='major', color='#AAAAAA', linestyle='-')
ax.minorticks_on()
ax.grid(b=True, which='minor', color='#AAAAAA', linestyle='-', alpha=0.2)
```

## GROWTH RATE (%)

```python
growth_rates = pd.DataFrame()
growth_rates['data'] = pd.to_datetime(national_trend['data']).dt.date
growth_rates_columns = ['nuovi_attualmente_positivi','ricoverati_con_sintomi', 'terapia_intensiva', 'totale_ospedalizzati', 'isolamento_domiciliare', 'totale_attualmente_positivi', 'dimessi_guariti', 'deceduti', 'totale_casi', 'tamponi']

for c in growth_rates_columns:
    growth_rates[c] = (national_trend[c] - national_trend[c].shift(1))/national_trend[c].shift(1)*100
growth_rates= growth_rates.round(2)
#growth_rates=growth_rates[1:]
```

```python
# %matplotlib notebook
plt.rcParams['figure.figsize'] = [20, 5]

colors=['purple','red','green','#00FF00','#0000FF','#3AFAF0','#000000']
#legends = ["Totale Casi", "Totale Positivi","Ricoverati","Terapia Intensiva","Isolamento domiciliare","Guariti","Deceduti"]

plot =['totale_casi','nuovi_attualmente_positivi', 'terapia_intensiva', 'totale_ospedalizzati']
colors=['purple','red','green','#00FF00','#0000FF','#3AFAF0','#000000']
ax = growth_rates.plot(kind='line', x='data', y=plot, color=colors, rot=90)
plot =['isolamento_domiciliare', 'dimessi_guariti', 'deceduti','tamponi']
colors=['#00FF00','#0000FF','#3AFAF0','#000000']
ay = growth_rates.plot(kind='line', x='data', y=plot, color=colors, rot=90)

for g in [ax,ay]:
    g.legend(loc="upper left")
    g.set_ylim(-20,150)
    g.set_xlabel('Date')
    g.set_ylabel('Casi')
    g.grid(b=True, which='major', color='#AAAAAA', linestyle='-')
    g.minorticks_on()
    g.grid(b=True, which='minor', color='#AAAAAA', linestyle='-', alpha=0.2)

```

## TREND NUOVI POSITIVI

```python
avg_trend = pd.DataFrame()
avg_trend['data'] = national_trend['data']
avg_trend['nuovi_attualmente_positivi_casi'] = national_trend['nuovi_attualmente_positivi']
avg_trend['nuovi_attualmente_positivi_percent'] = growth_rates['nuovi_attualmente_positivi']
avg_trend['avg_nuovi'] = growth_rates['nuovi_attualmente_positivi'].rolling(7).mean().round(2)
avg_trend['tamponi'] = incremental_data['tamponi']
avg_trend['tamponi_percent'] = growth_rates['tamponi']
avg_trend['avg_tamponi'] = growth_rates['tamponi'].rolling(7).mean().round(2)
avg_trend['deceduti'] = incremental_data['deceduti']
avg_trend['deceduti_percent'] = growth_rates['deceduti']
avg_trend['avg_deceduti'] = growth_rates['deceduti'].rolling(7).mean().round(2)
```

```python
# %matplotlib notebook
plt.rcParams['figure.figsize'] = [20, 5]

plot = ['nuovi_attualmente_positivi_percent','avg_nuovi']
colors = ['purple','green','blue', 'red']
labels = ['Growth Rate Nuovi Casi', 'Media 7 giorni']
ax = avg_trend.plot(kind='line', x='data', y=plot, color=colors, rot=90, label=labels)
ax.set_ylim(-20,50)
ax.set_title("Nuovi Casi")



plot = ['tamponi_percent', 'avg_tamponi']
colors = ['blue', 'red']
labels = ['Growth Rate Nuovi Tamponi', 'Media 7 giorni']
ay = avg_trend.plot(kind='line', x='data', y=plot, color=colors, rot=90, label=labels)
ay.set_ylim(0,30)
ay.set_title("Nuovi Tamponi")

plot = ['deceduti_percent', 'avg_deceduti']
colors = ['black', 'gray']
labels = ['Growth Rate Deceduti', 'Media 7 giorni']
az = avg_trend.plot(kind='line', x='data', y=plot, color=colors, rot=90, label=labels)
az.set_ylim(0,50)
az.set_title("Deceduti")

for g in [ax, ay, az]:
    g.legend(loc="upper left")
    
    g.set_xlabel('Date')
    g.set_ylabel('Percent')
    g.grid(b=True, which='major', color='#AAAAAA', linestyle='-')
    g.minorticks_on()
    g.grid(b=True, which='minor', color='#AAAAAA', linestyle='-', alpha=0.2)
```

```python
percents_in_period = pd.DataFrame()
percents_in_period['data'] = national_trend['data']
percents_in_period['ricoverati_con_sintomi'] = (national_trend['ricoverati_con_sintomi'] / national_trend['totale_casi'])*100
percents_in_period['terapia_intensiva'] = (national_trend['terapia_intensiva'] / national_trend['totale_casi'])*100
percents_in_period['isolamento_domiciliare'] = (national_trend['isolamento_domiciliare'] / national_trend['totale_casi'])*100
percents_in_period['dimessi_guariti'] = (national_trend['dimessi_guariti'] / national_trend['totale_casi'])*100
percents_in_period['deceduti'] = (national_trend['deceduti'] / national_trend['totale_casi'])*100
percents_in_period['contagi'] = growth_rates['nuovi_attualmente_positivi']
percents_in_period = percents_in_period[1:].round(2)
```

```python
# %matplotlib notebook
plt.rcParams['figure.figsize'] = [20, 5]

plot = ['ricoverati_con_sintomi','terapia_intensiva','isolamento_domiciliare','dimessi_guariti','deceduti']
colors = ['purple','green','blue', 'orange','red']
labels = ['ricoverati_con_sintomi','terapia_intensiva','isolamento_domiciliare','dimessi_guariti','deceduti']
ax = percents_in_period.plot(kind='bar', x='data', y=plot, color=colors, rot=90, label=labels, stacked=True)
ax.set_title("Dispesione percentuale")
ax.set_ylim(0,100)

plot = ['contagi']
colors = ['green']
labels = ['contagi']
ay = percents_in_period.plot(kind='bar', x='data', y=plot, color=colors, rot=90, label=labels)
ay.set_title("Contagi Growth Rate")
ay.set_ylim(-20,150)



for g in [ax,ay]:
    g.legend(loc="upper left")
    g.set_xlabel('Date')
    g.set_ylabel('Percent')
    g.grid(b=True, which='major', color='#AAAAAA', linestyle='-')
    g.minorticks_on()
    g.grid(b=True, which='minor', color='#AAAAAA', linestyle='-', alpha=0.2)
```
